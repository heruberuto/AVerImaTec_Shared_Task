```python
import json
# load /mnt/data/factcheck/averimatec/submissions/submission.json
submission = json.load(open("/mnt/data/factcheck/averimatec/submissions/submission.json", "r"))
```

```python
submission[1]
```

```python
for el in submission:
    for i, ev in enumerate(el["evidence"]):
        ev["text"] = el["questions"][i] + " " + ev["text"]
        if len(ev["images"]):
            ev["text"] += " [IMG_1]"
```

```python
# save pretty to submission_concat.json
with open("/mnt/data/factcheck/averimatec/submissions/submission_concat.json", "w") as f:
    json.dump(submission, f, indent=4)
```
